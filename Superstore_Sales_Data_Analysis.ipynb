<h1>Which Category of Product generate more Profit

```python
cf = df.groupby('Category',as_index=False)['Profit'].sum()
cf
```

```python
cf = df.groupby('Category',as_index=False)['Profit'].sum()
cf
sns.barplot(data = cf, x= 'Category',y = 'Profit',hue = 'Category')
plt.show()
#Technology products have the highest profit margin
```

<h1>Which Region has the Highest Sales

```python
rs = df.groupby('Region',as_index= False)['Sales'].sum()
rs
```

```python
rs = df.groupby('Region',as_index= False)['Sales'].sum()
rs
sns.catplot(data = rs, x= 'Region',y = 'Sales',hue = 'Region',kind = 'bar')
plt.xticks(rotation=90)
plt.show()
#The CENTRAL Region has the most Highest Sales
```

<h1>Higher Sales Months of the Year

```python
df['Order_Month'] = pd.DatetimeIndex(df['Order Date']).month
ms = df.groupby('Order_Month',as_index= False)['Sales'].sum()
ms
```

```python
df['Order_Month'] = pd.DatetimeIndex(df['Order Date']).month
ms = df.groupby('Order_Month',as_index= False)['Sales'].sum()
ms
sns.relplot(data = ms,kind= 'scatter',x = 'Order_Month',y = 'Sales',hue = 'Order_Month',markers= 'O',alpha = 1)
plt.show()
#Sales are Higher in NOVEMBER and DECEMBER
```

<h1>Lowest Rate of Returned Products

*   Total Shipping Orders
*   Returned Shipping Orders

*   Average Returned Orders as Per Class






```python
#Total Ship Counts by Shipping Mode
total_ship_counts = df['Ship Mode'].value_counts().reset_index()
total_ship_counts.columns = ['Ship Mode', 'Total Orders']
total_ship_counts.head()
```

```python
sns.barplot(data = total_ship_counts , x= 'Ship Mode',y = 'Total Orders',hue = 'Ship Mode')
plt.show()
```

```python
#Returned Shipping Orders
returned_order_ship_mode = df[df['Profit'] < 0]['Ship Mode'].value_counts().reset_index()

returned_order_ship_mode.columns = ['Ship Mode', 'Returned Orders']
returned_order_ship_mode
```

```python
sns.barplot(data = returned_order_ship_mode,x = 'Ship Mode',y = 'Returned Orders',hue = 'Ship Mode')
plt.show()
```

```python
# calculate return percentage
return_percentage_every_mode = (returned_order_ship_mode / total_ship_counts * 100).reset_index()
return_percentage_every_mode.columns = ['Ship Mode', 'Returned Percentage']

return_percentage_every_mode
```

```python
sns.barplot(data = return_percentage_every_mode,x = 'Ship Mode',y = 'Returned Percentage',hue = 'Ship Mode')
plt.show()
```

<h1> Check the Company Profit in Weakdays and Weakends

```python
#create new column as Order day
df['Order Day'] = pd.DatetimeIndex(df['Order Date']).day_name()
df['Order Day'].head(3)
```

```python
#Per Day Profit
day_profit = df.groupby('Order Day')['Profit'].sum().reset_index()
day_profit.columns =['Day','Profit']
day_profit
```

```python
sns.barplot(data = day_profit, x ='Day', y= 'Profit',hue = 'Day')
plt.show()
```
